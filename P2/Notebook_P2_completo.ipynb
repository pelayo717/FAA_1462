```python
""" 
Hemos puesto a prueba el clasificador KNN, usando como K valores de 1,5, 11 y 21 a la par que probamos las diferentes funciones de distancia (Euclidea, Manhattan y Mahalanobis). Sabemos que este tipo de clasificador es muy ventajoso en cuanto a la simplicidad de implementación, pero como desventaja principal, vemos que es más util en datasets de pocas instancias puesto que consume muchos recursos de CPU y memoria. A su vez hemos probado las distintas configuraciones
sin normalizar y normalizando aquellos atributos con datos decimales. Esto es debido a que, el clasificador, tiende a dar más imporancia a aquellos datos que no han sido normalizados. Es decir, estamos facilitando el trabajo al clasificador, para que este sea lo más preciso y realista en sus predicciones.

Antes de evaluar cada configuracion, debemos especificar la peculiaridad de los k's escogidos asi como de las funciones seleccionadas. El hiperparámetro 'k' indica el número de ejemplos cercanos al ejemplo que intentamos clasificar en los que nos vamos a fijar para predecir la clase del test.

El uso de K en este algoritmo puede ser decisivo, de cara a los resultados obtenidos. Un K excesivamente pequeño (del orden de 1) es util en cuanto a datasets donde el espectro muestral presenta grandes diferencias encuanto a unas instancias de otras. Si nosotros simplificamos el problema, y lo llevamos a un entorno donde los atributos albergan valores nominales (enteros pero no decimales) veremos que existe una mayor probabilidad de que los puntos sean más facilmente 
separables y por ende no sea necesario usar un K superior a uno.

Dado que, esto no es la situacion anterior, y nuestros datasets son mas complejos, consideramos oportuno dar valores más grandes de K. Cuanto mayor sea, menos ruido estaremos introduciendo en nuestro clasificador, y con ello conseguiremos unas clasificaciones mucho más realistas. Es necesario mencionar, que al tener en cuenta más puntos respecto a la instancia evaluada la complejidad computacional aumentará.

En cuanto a las funciones empleadas, conocemos la peculiaridad de que la distacia Euclídea (Lasso 1) es peor en cuanto a datasets pequeños respecto a Manhattan (Lasso 2) que es capaz de manejar grandes cantidades de datos. Por otro lado, utilizar la funcion de Mahalanobis nos va a permitir una correlación entre los atributos del dataset. Esta caracteristica propia de esta función permite, no simplificar el uso de los datos como hacen distancia Euclidea y Manhattan. 

Si nos fijamos en los resultados obtenidos mediante el cálculo con distancia Euclídea, vemos que se confirma que la normalización de los datos facilita mucho el trabajo de este clasificador y hace que sus predicciones sean mucho más precisas. 

Centrándonos ahora en la distancia de Manhattan, ocurre un suceso interesante y es que para el dataset 'WDBC' es mejor normalizar los datos mientras que para 'Diabetes' no. Esto se podría deber a que las diferencias de magnitudes entre los datos de 'Diabetes' no son tan grandes como las de 'WDBC' y además que en el primero sólamente se deben normalizar dos atributos (por ser decimales), mientras que en este último se deben normalizar todos.

Por último, con la distancia de Mahalanobis no se encuentra ninguna diferencia entre normalizar los datos o no hacerlo. Esto es porque al calcular las distancia entre dos ejemplos, también se calcula con el conjunto de datos haciendola así invariante ante cambios de escala.

En cuanto al K del dataset 'Diabetes', vemos que el que mejores resultados nos proporciona es 21. Esto nos indica que los ejemplos se encuentran, como hemos comentado antes, bastante repartidos en el espacio muestral. Lo que se quiere decir que no hay zonas donde se distinga especialmente la pertenencia de los ejemplos a una clase o a otra.

En cuanto al K del dataset 'WDBC', vemos que el que mejores resultados nos proporciona, en la media de las distancias, es el 5. Por lo que, de forma contraria al dataset anterior, en este los ejemplos sí que están repartidos de forma que se puedan distinguir diferentes zonas en el espacio muestral. 

Ambos K's se explicarán también más adelante en la parte del Espacio ROC (Apartado 4).

"""
```

```python
""" 
Del mismo modo que hicimos con KNN, ahora estudiaremos la importancia que tienen los hiperparamteros con los que hemos
configurado el clasificador de regresion logistica.

Debemos tener en cuenta el que hemos empeleado una tasa de aprendizaje y un numero de epocas de entrenamiento como
hiperparametros, aunque la regresión logistica, casi como cualquier otro clasificador recoge otros muchos posibles valores
que permiten realizar distintos tipos de configuraciones atendiendo a unas necesidades o a unos contextos concretos.

En cuanto a la tasa (o ratio) de aprendizaje, su funcion es garantizar que W, se ma modificando con una rapidez 
mas o menos comprensile y buscando siempre minimizar la funcion de coste. Es decir, de escoger un valor demasiado
grande de tasa, estaremos provocando cambios muy bruscos en las componentes de este vector. En caso contrario,
estaremos asegurandonos que este cambio sera muy paulatino, y posiblemente encuentre correctamente una funcion que
devuelva un error minimo. Sine embargo, ello tambien implica la lentitud con la que se llega a esa configuracion deseada.

Esto son condiciones muy apreciables, cuando se añade al clasificador, un error que debe superar por debajo. 
Cuando se emplean un error minimo alcanzable es recomendable usar un gran numero de epocas de entrenamiento, para poder
tumbar ese error lo maximo posible. Es entonces cuando, epoca tras epoca ese error generado por nuestra funcion busca
ser menor al que nosotros introdujimos. Esto puede provocar largos tiempos de entrenamiento, pues el error de nuestro
clasificador puede nunca llegar a ser inferior. En caso contrario, es una forma de que al cumplir esa hito marcado
, el clasificador deje de entrenar y con ello evitemos consumir recursos de manera innecesaria.

Por otra parte encontramos la necesidad de escoger un correcto numero de epocas. Como comentamos en el parrafo anterior
puede ser una forma decisiva de minimizar nuestra funcion de coste, asi como una forma descabellada de malgastar 
recursos cuando ese coste no es posible minimizarlo mas.

Para poner a prueba nuestro clasificador de regresión logística, hemos probado a con 100, y 500 épocas, asi como con 0,5
y 0,75 de tasa de aprendizaje. Podemos observar, como aquella configuración que presenta un mejor resultado en cuanto a,
Diabetes se refiere, es empleando validación cruzada con la segunda, tercera y cuarta configuración. Sin embargo,
si prestamos atención veremos como la primera configuración (100 y 0.5) es la que mejor resultado devuelve para Diabetes 
en validación simple. Consideramos que al no emplear una condición de error durante el entrenamiento, 
se ha conseguido una función de coste que puede ser bastante mejorable usando un número mayor de épocas (del orden del millar)
y puede que reduciendo la tasa de aprendizaje lo máximo posible.

No obstante esta situación es totalmente contraria en validación cruzada donde no parece que se este sobreajustando
y es por que ello que el resultado de las 3 últimas configuraciones sea tan similar. Esto es un claro indicador, de que
estos clasificadores pueden ser mejorables, hasta el momento en el que una de las tres configuraciones destaque frente
a las demás.

En cuanto a Wdbc, en general su error es mucho menor. Esto es probable que suceda, debido al mayor número de atributos con los que cuenta.
Debemos recordar que estos no son atributos categóricos, sino decimales. Con ello, debemos tener en cuenta que existe
una mayor facilidad de clasificación y por ende, un mejor resultado. Si profundizamos, veremos como aquella configuración
que mejor reusltado aporta es empleando validación cruzada, y usando tanto la configuración 2,3 y 4s.

Es curioso, como tanto para Diabetes como para Wdbc, no conseguimos sacar aquella que nos devuelve un resultado, ligeramente
diferente respecto las demás, sino que estamos viendo cual de ellas no es una configuración lo suficientemente buena como
para tenerla en cuenta. En general, la configuración 2,3, y 4 son similares y sus errores también. 

Podemos conluir que la primera configuración es la que mejores valores restorna (para ambos ficheros), pero solo con validación
simple. En el resto de puntuaciones son las demás configuraciones las que salen victoriosas.

"""
```
